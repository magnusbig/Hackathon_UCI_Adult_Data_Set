```python
import pandas as pd
```

## Training Data

```python
train_1 = pd.read_csv('../data/clean_train_1.csv')
train_2 = pd.read_csv('../data/clean_train_2.csv')
```

```python
train_1.head()
```

```python
train_2.head()
```

```python
# concatenate
train_clean = pd.concat([train_1,train_2], axis=1)
```

```python
# Save to csv
train_clean.to_csv('../data/clean_train.csv', index=False)
```
